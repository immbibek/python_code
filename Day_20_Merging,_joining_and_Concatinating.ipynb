```python
import numpy as np
import pandas as pd

```

**Concat() Function**

```python
courses=pd.read_csv('/content/courses (1).csv')
delivery=pd.read_csv('content/deliveries (1).csv')
nov=pd.read_csv('/content/reg-month1.csv')
dec=pd.read_csv('/content/reg-month2.csv')
students=pd.read_csv('/content/students.csv')
matches=pd.read_csv('/content/matches - matches.csv')
```

```python
# pd.concat

#ingnore_index
#df.append
# multiindex --> fetch using iloc
# concat dataFrame horizontally
```

```python
regs=pd.concat([nov,dec],ignore_index=True)
regs
```

```python
# concat DataFrame vertically
multi=pd.concat([nov,dec],keys=['Nov','Dec'])   #--> Multiindex DataFrame
```

```python
multi
```

```python
multi.loc['Nov',0]
```

```python
# if I want concat Horizontally
pd.concat([nov,dec],axis=1)
```

**Merging**

```python
# inner join
students.merge(regs,how='inner',on='student_id')           # merg on the basis of common in both dataframe
```

```python
# left join
courses.merge(regs,how='left',on='course_id')
```

```python
# just join temp Dataframe to main
temp_df=pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit',"Rahul"],
    'partner':[28,26,17]

})
pd.concat([students,temp_df],ignore_index=True)
```

```python
# right
students.merge(regs,how='right',on='student_id')
```

```python
# outer join
students.merge(regs,how='outer',on='student_id')
```

```python
# 1. Find Total Revnue
regs.merge(courses,how='inner',on='course_id')['price'].sum()
```

```python
# 2.find month by month revnue
temp_df=pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
temp_df.merge(courses,on='course_id').groupby('level_0')['price'].sum()
```

```python
# print the registration table
# cols --> name --> course--> price

regs.merge(students,on='student_id').merge(courses,on='course_id')[['name','course_name','price']]
```

```python
# plot bar charts for revnue/course
regs.merge(courses,on='course_id').groupby('course_name')['price'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
# 5.find students who enrolled in both the months

common_student_id=np.intersect1d(nov['student_id'],dec['student_id'])
common_student_id
```

```python
students[students['student_id'].isin(common_student_id)]
```

```python
# 6. find course that got no enrollement
#courses['course_id']
# regs['course_id']
course_id_list=np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(course_id_list)]
```

```python
# 7.find students who did not enroll into any courses
students_id_list=np.setdiff1d(students['student_id'],regs['student_id'])
students[students['student_id'].isin(students_id_list)]
```

```python
# 8. print student name --> partner name for all anrolled students
# self join
students.merge(students,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```

```python
# 9.find top 3 students who did most number enrollments
regs.merge(students,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# 10. find top 3 students who spent most amount of money on courses
regs.merge(students,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```
